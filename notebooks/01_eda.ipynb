# Purpose of EDA Notebook:

# Credit Card Fraud Detection - EDA

# 1. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Set plotting styles
```

```python
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10,6)
```

# 2. Load Dataset

```python
data_path = '../data/raw/creditcard.csv'  
df = pd.read_csv(data_path)
```

```python
df
```

# 3. Initial Data Inspection

```python
print("Dataset shape:", df.shape)
print("\nDataset Info:")
print(df.info())
print("\nFirst 5 rows:")
display(df.head())
```

# 4. Check Missing Values

```python
missing = df.isnull().sum()
print("\nMissing Values in Each Column:\n", missing[missing > 0])
```

# 5. Check Target Distribution (Class Imbalance)

```python
print("\nTarget Value Counts:")
print(df['Class'].value_counts())
```

```python
#Visualize Class Imbalance
```

```python
sns.countplot(x='Class', data=df)
plt.title("Class Distribution (0 = Non-Fraud, 1 = Fraud)")
plt.show()
```

# 6. Summary Statistics

```python
display(df.describe())
```

# 7. Correlation Analysis

```python
corr_matrix = df.corr()
```

```python
corr_matrix
```

```python
#Heatmap

plt.figure(figsize=(12,10))
sns.heatmap(corr_matrix, cmap='coolwarm_r', annot=False, linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```

```python
#Focus on correlation with target

target_corr = corr_matrix['Class'].sort_values(ascending=False)
print("\nTop features correlated with target:\n", target_corr.head(10))
```

# 8. Visualize Important Features

```python
#Amount Distribution by Class

sns.boxplot(x='Class', y='Amount', data=df)
plt.title("Transaction Amount by Class")
plt.show()
```

```python
#Time Distribution by Class

sns.histplot(data=df, x='Time', hue='Class', bins=50, kde=True)
plt.title("Transaction Time Distribution by Class")
plt.show()
```

# 9. Feature Distribution Check

```python
#Visualize first 5 PCA features (V1-V5)

pca_features = [f'V{i}' for i in range(1,6)]
df[pca_features].hist(bins=50, figsize=(12,8))
plt.suptitle("Distribution of PCA Features V1-V5")
plt.show()
```

# 10. Outlier Analysis

```python
#Amount Outliers

sns.boxplot(df['Amount'])
plt.title("Transaction Amount Outliers")
plt.show()
```

# 11. Insights Summary

```python
print("""
EDA Insights:
1. Dataset is highly imbalanced: 0 = Non-Fraud, 1 = Fraud (~0.17% frauds)
2. 'Amount' shows skewed distribution; scaling will be required.
3. Most features are anonymized PCA features (V1-V28)
4. Correlation with 'Class' shows V14, V17, V12, V10 are more correlated with fraud
5. Time distribution does not show clear pattern but can be used as feature
6. No missing values detected
7. Outliers exist in 'Amount', consider log-scaling or robust scaling
""")
```
